```python
import numpy as np
```

```python
ppDF['distance_from_middle_of_plate'] = np.abs(ppDF['x0'])
```

```python
ppDF.head()
```

```python
ppDF['batter'].str.replace('.',' ').sample(5)
```

```python
(ppDF['pitcher'] + 'throws a' + ppDF['pitch_type'] + ' to ' + ppDF['batter']).sample(5)
```

```python
def is_fastball(pitch):
    """
    takes a string named pitch ('CH','FF','FC',etc) and checks whether it's a fastball
    (cutter,four-seam,'two-seam',sinker,splitter)
    """
    return pitch in ['FC','FS','FT','SI']
```

```python
ppDF['is_fastball'] = ppDF['pitch_type'].apply(is_fastball)
#.apply() takes a function as an argument and applies to each value in a column
ppDF[['pitcher','is_fastball','pitch_type','mph']].sample(5)
```

```python
ppDF['is_fastball_lambda'] = ppDF['pitch_type'].apply(lambda x: x in ['FC','FS','FT','SI'])
```

```python
ppDF.head()
#can find our lambda column in dataframe here
```

```python
ppDF.drop('is_fastball_lambda',axis=1,inplace=True)
```
