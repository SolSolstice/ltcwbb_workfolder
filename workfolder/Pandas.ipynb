```python
players = players.set_index('playerID')
players.head()
```

```python
players_cuba = players.loc[players['birthCountry']=='Cuba',['name','height','weight','birthCountry']]
players_cuba.head()
```

```python
players_cuba.sort_values('weight',inplace=True)
players_cuba.head()
```

```python
players_cuba['bats'] = players['bats']
players_cuba
```
